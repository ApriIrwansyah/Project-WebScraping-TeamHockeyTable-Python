```python
rows = soup.find_all('div', {'class' : 'row'})[2]
for row in rows.find_all('p'):
    print(row.get_text())
    # if row.find_all('p') == [0]:
    #     print(row.get_text())
    # elif row.find_all('p') == [1]:
    #     print(row.get_text())
# row.text
```

```python
table   = soup.find_all('table', {'class' : 'table'})[0]
# table
rows    = table.find_all('tr')
# rows
for row in rows:
    # print(row.get_text())
    for item in row.find_all(['td', 'th']):
        print(item.get_text())
```

```python
hasil = []
for row in rows:
    info = []
    for item in row.find_all(['td', 'th']):
        # print(item.get_text())
        info.append(item.text
                    .replace('\n                            ', '')
                    .replace('\n                        ', ''))
    hasil.append(info)
print(hasil)
```

```python
import pandas as pd
df = pd.DataFrame(hasil) # Convert list kedalam DataFrame
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header # mengubah first row menjadi column header
df.reset_index(drop=True, inplace=True) #reset index sehingga dimulai dari 0
display(df)
```

```python
df.to_csv('hockeyTeam.csv')
df.to_excel('hockeyTeam.xlsx', sheet_name='hockeyTeam')
df.to_json('hockeyTeam.json')
      
```
