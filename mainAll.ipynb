```python
# first
hasil = []
for row in first_100:
    info = []
    for item in row.find_all(['td', 'th']):
        # print(item.get_text())
        info.append(item.text
                    .replace('\n                            ', '')
                    .replace('\n                        ', ''))
    hasil.append(info)
print(hasil)

# print(len(first_100))
```

```python
hasil = []
for pageNumber in range(1, 7):
    url = 'https://www.scrapethissite.com/pages/forms/?page_num=1&per_page=100'

    # ditahap ini kita akan menyatukan beberapa halaman
    # Cara 1
    # for i in range(1, 50):
    #     urls = 'https://www.scrapethissite.com/pages/forms/?page_num={}&per_page=100'.format(i)
    # print(urls)

    # Cara 2
    urls = f'https://www.scrapethissite.com/pages/forms/?page_num={pageNumber}&per_page=100' # .format(pageNumber)
    resp = requests.get(urls)
    soup = BeautifulSoup(resp.content, 'html.parser')
    print(urls)
    # first_600 = soup.find_all('table', {'class' : 'table'})
    for row in first_100:
        item = {}
        # kita akan mengambil per items
        item['Team Name']       = row.find('td', {'class' : 'name'}).get_text().replace('\n                            ', '').replace('\n                        ', '') 
        item['Year']            = row.find('td', {'class' : 'year'}).get_text().replace('\n                            ', '').replace('\n                        ', '')
        item['Wins']            = row.find('td', {'class' : 'wins'}).get_text().replace('\n                            ', '').replace('\n                        ', '') 
        item['Losses']          = row.find('td', {'class' : 'losses'}).get_text().replace('\n                            ', '').replace('\n                        ', '') 
        item['OT Losses']       = row.find('td', {'class' : 'ot-losses'}).get_text().replace('\n                            ', '').replace('\n                        ', '') 
        item['Win %']           = row.find('td', {'class' : 'pct'}).text.replace('\n                            ', '').replace('\n                        ', '')
        item['Goals For (GF)']  = row.find('td', {'class' : 'gf'}).get_text().replace('\n                            ', '').replace('\n                        ', '')
        item['Goal Against (GA)']   = row.find('td', {'class' : 'ga'}).get_text().replace('\n                            ', '').replace('\n                        ', '')
        item['+/-']             = row.find('td', {'class' : 'diff'}).get_text().replace('\n                            ', '').replace('\n                        ', '')
        hasil.append(item)
        
        # for item in row.find_all(['td', 'th']):
        #     # print(item.get_text())
        #     info.append(item.text
        #                 .replace('\n                            ', '')
        #                 .replace('\n                        ', ''))
        # hasil.append(info)
    print(hasil)

    # print(soup)
```

```python
df = pd.DataFrame(hasil) # Convert list kedalam DataFrame
df.to_excel('HockeyTeam600.xlsx', sheet_name='Team')
```
